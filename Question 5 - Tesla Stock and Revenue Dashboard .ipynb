```python
import yfinance as yf
import pandas as pd
```

```python
tesla = yf.Ticker("TSLA")
```

```python
tesla_stock_data = tesla.history(period = "5y")
```

```python
print(tesla_stock_data.head())
```

```python
csv_file = 'tesla_stock_data.csv'
tesla_stock_data.to_csv(csv_file)
```

```python
load_data =pd.read_csv(csv_file, index_col='Date', parse_dates=True)
```

```python
print(load_data.head())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.plot(load_data.index, load_data['Close'], label='Close Price')
plt.title('Tesla Stock Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
