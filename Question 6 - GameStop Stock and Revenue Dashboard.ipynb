```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
GameStop = yf.Ticker("GME")
GameStop_stock_data = GameStop.history(period="5y")
```

```python
csv_file = 'GameStop_stock_data.csv'
GameStop_stock_data.to_csv(csv_file)
```

```python
loaded_data = pd.read_csv(csv_file, index_col='Date', parse_dates=True)
```

```python
# print(loaded_data.head())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(loaded_data.index, loaded_data['Close'], label='Close Price')
plt.title('GameStop Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
